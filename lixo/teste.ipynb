```python
# pipeline_tempo.py
import os
import pandas as pd
from datetime import datetime, timedelta
from urllib.parse import quote
from dotenv import load_dotenv



import requests
import pandas as pd
from io import StringIO

def extrair_dados(cidade, dias=7):
    
    data_inicio = datetime.today()
    data_fim = data_inicio + timedelta(days=dias)
    data_inicio = data_inicio.strftime('%Y-%m-%d')
    data_fim = data_fim.strftime('%Y-%m-%d')
    
    load_dotenv()  # Busca .env na raiz automaticamente
    VISUAL_CROSSING_KEY = os.getenv('VISUAL_CROSSING_KEY')
    
    city_encoded = quote(cidade)
    url = f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{city_encoded}/{data_inicio}/{data_fim}?unitGroup=metric&include=days&key={VISUAL_CROSSING_KEY}&contentType=csv'

    r = requests.get(url, timeout=10)
    print(r.status_code)

    df = pd.read_csv(StringIO(r.text))
    return df

extrair_dados('Caxias do Sul', 7)
```

```python
# pipeline_tempo_test.py
import os
import pandas as pd
from datetime import datetime, timedelta
from urllib.parse import quote
from dotenv import load_dotenv
import requests
from io import StringIO
import time

def extrair_dados(cidade, dias=7):
    """Extrai dados da API Visual Crossing"""
    
    data_inicio = datetime.today()
    data_fim = data_inicio + timedelta(days=dias)
    data_inicio = data_inicio.strftime('%Y-%m-%d')
    data_fim = data_fim.strftime('%Y-%m-%d')
    
    load_dotenv()  # Busca .env na raiz
    VISUAL_CROSSING_KEY = os.getenv('VISUAL_CROSSING_KEY')
    
    city_encoded = quote(cidade)
    url = f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{city_encoded}/{data_inicio}/{data_fim}?unitGroup=metric&include=days&key={VISUAL_CROSSING_KEY}&contentType=csv'

    r = requests.get(url, timeout=10)
    r.raise_for_status()  # Erro se falhar
    df = pd.read_csv(StringIO(r.text))
    return df

# --- Loop minuto a minuto ---
while True:
    try:
        df = extrair_dados("Caxias do Sul", dias=7)
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        path = f"./previsao_tempo_{timestamp}.csv"
        df.to_csv(path, index=False)
        print(f"✅ Dados salvos em {path} | {datetime.now()}")
    except Exception as e:
        print("❌ Erro:", e)
    
    # Espera 60 segundos antes da próxima requisição
    time.sleep(60)
```
